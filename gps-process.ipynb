```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyproj import Proj

%matplotlib inline
```

```python
myProj = Proj("+proj=utm +zone=23K, +south +ellps=WGS84 +datum=WGS84 +units=m +no_defs")
data = pd.read_csv("record.csv", sep=";")
```

```python
data.head()
```

```python
plt.plot(data.lon, data.lat)
```

```python
UTMx, UTMy = myProj(list(data.lon), list(data.lat))
```

```python
plt.plot(UTMx, UTMy)
```

```python
UTMx, UTMy = np.array(UTMx), np.array(UTMy)
```

```python
print(UTMx.min(), UTMx.max())
```

```python
utmx = UTMx - (UTMx.min() - 20)
utmy = UTMy - (UTMy.min() - 20)
```

```python
plt.plot(utmx, utmy)
```

```python
print((-utmx.min()+utmx.max())/2)
print((-utmy.min()+utmy.max())/2)
```
